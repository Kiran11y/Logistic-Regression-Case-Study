### 5A. Splitting The Data 

```python
y = df.TenYearCHD
```

```python
x = df.iloc[:,:-1]
```

```python
##CODE HERE
X_cat = df[['male',"education","currentSmoker","prevalentStroke","prevalentHyp","diabetes"]]

X_cat.head()
```

```python
##CODE HERE
X_num = df[["age","cigsPerDay","BPMeds","totChol","sysBP","diaBP","BMI","heartRate","glucose"]]
X_num.head()
```

### 5B. Apply Data Transformations.

Hint: Use Standardization technique or Normalization technique

```python
## CODE HERE
X_concat = pd.concat([X_cat, X_num], axis=1)

X_concat.head()
```

### 5C. Split The Data into Train and Test

```python
##CODE HERE
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_concat, y, test_size = 0.30, random_state = 143)
```

```python
X_train.head()
```

```python
X_train_num = X_train.iloc[:,6:]
X_train_num.head()
```

```python
X_train_cat = X_train.iloc[:,:6]
X_train_cat.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_num_scaled = pd.DataFrame(scaler.fit_transform(X_train_num), 
                                    columns = X_train_num.columns, 
                                    index = X_train_num.index)
x_num_scaled.head()
```

```python
X_transformed_train = pd.concat([X_train_cat,x_num_scaled], axis=1)

X_transformed_train.head()
```

### for X_test

```python
X_test.shape
```

```python
X_test_num = X_test.iloc[:,6:]
X_test_num.head()
```

```python
X_test_cat = X_test.iloc[:,:6]
X_test_cat.head()
```

```python
from sklearn.preprocessing import StandardScaler
x_num_test = pd.DataFrame(scaler.transform(X_test_num), 
                                    columns = X_test_num.columns, 
                                    index = X_test_num.index)
x_num_test.head()
```

```python
X_transformed_test = pd.concat([X_test_cat,x_num_test], axis=1)

X_transformed_test.head()
```

### 6A. Apply The Logistic Regression Model.

```python
##CODE HERE
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(X_transformed_train, y_train)
```

```python
y_pred = classifier.predict(X_transformed_test)
y_pred
```

```python
##CODE HERE
```

### 6B.Find best threshold value using Log-Loss 

```python
from sklearn.metrics import roc_auc_score, roc_curve 
```

```python
## CODE HERE
pred = []

for model in [classifier]:
    pred.append(pd.Series(model.predict_proba(X_transformed_test)[:,1]))

final_pred = pd.concat(pred, axis=1).mean(axis=1)
print(" test roc-auc:{}".format(roc_auc_score(y_test,final_pred))) 
False_pos_rate, True_pos_rate, threshold = roc_curve(y_test, final_pred)

threshold
```

### Accuracy

```python
##CODE HERE
from sklearn.metrics import accuracy_score

acc = []

for thres in threshold:
    y_pred = np.where(final_pred>thres,1,0)

    #what ever prediction i am getting and if it is greater than threshold i'll be converting as 1 or i'll keep it as 0.




    acc.append(accuracy_score(y_test,y_pred,normalize=True))

    #Then i'll be computing my accuracy score with my y_test and then
    #append the accuracy inside acc list.



acc = pd.concat([pd.Series(threshold), pd.Series(acc)], axis=1)
acc.columns = ['threshold','accuracy']
acc.sort_values(by="accuracy", ascending=False, inplace = True)
acc.head()

```

```python
import matplotlib.pyplot as plt
def plot_roc_curve(False_pos_rate,True_pos_rate):
    plt.plot(False_pos_rate, True_pos_rate, label="ROC")
    plt.plot([0,1],[0,1],color="Red",linestyle="--")
    plt.xlabel("False_Positive_rate")
    plt.ylabel("True_Positive_rate")
    plt.title("ROC Curve")
    plt.legend
    plt.show()

plot_roc_curve(False_pos_rate,True_pos_rate) 
```
